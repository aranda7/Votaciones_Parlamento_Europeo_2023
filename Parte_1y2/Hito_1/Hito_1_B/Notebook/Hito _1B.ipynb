## Justificaci√≥n Acad√©mica de los Modelos Seleccionados 

### Introducci√≥n

La elecci√≥n de los modelos en un proyecto de Ciencia de Datos es una decisi√≥n de extrema importancia que debe ser rigurosamente justificada, especialmente en el contexto de un trabajo de fin de m√°ster. Esta justificaci√≥n no solo valida la metodolog√≠a empleada, sino que tambi√©n demuestra una comprensi√≥n profunda de las capacidades y limitaciones de los algoritmos de Machine Learning en relaci√≥n con el problema de investigaci√≥n. En este documento, se presenta la justificaci√≥n acad√©mica para la selecci√≥n de los siguientes modelos de clasificaci√≥n: Regresi√≥n Log√≠stica, Random Forest Classifier, Gradient Boosting y Perceptr√≥n Multicapa (MLP) / Red Neuronal. Cada uno de estos modelos ha sido elegido estrat√©gicamente para ofrecer una visi√≥n metodol√≥gica integral, permitiendo una evaluaci√≥n exhaustiva del problema y la optimizaci√≥n de los resultados.

La metodolog√≠a de selecci√≥n se basa en la b√∫squeda de un equilibrio entre la interpretabilidad, la robustez, el rendimiento predictivo y la capacidad de explorar diferentes paradigmas de aprendizaje autom√°tico. Desde un modelo lineal de referencia hasta t√©cnicas de ensamble avanzadas y una incursi√≥n en el aprendizaje profundo, la combinaci√≥n de estos algoritmos permite no solo identificar el modelo m√°s adecuado para la tarea, sino tambi√©n comprender las caracter√≠sticas intr√≠nsecas de los datos y la complejidad de las relaciones subyacentes. Esta aproximaci√≥n sistem√°tica asegura que las conclusiones derivadas del an√°lisis sean s√≥lidas, fiables y acad√©micamente defendibles.



### Justificaci√≥n Acad√©mica de los Modelos Seleccionados

Para la selecci√≥n de modelos en este proyecto, se ha optado por un conjunto de algoritmos de clasificaci√≥n que, en su conjunto, ofrecen una visi√≥n metodol√≥gica integral del problema. Cada modelo aporta una contribuci√≥n espec√≠fica y es fundamental para la fase de comparaci√≥n y optimizaci√≥n.

#### 1. Regresi√≥n Log√≠stica (Logistic Regression) üìä

**Contribuci√≥n Metodol√≥gica:** La Regresi√≥n Log√≠stica se incluye como el modelo de l√≠nea base (baseline) lineal fundamental. Su funci√≥n principal es establecer un umbral de rendimiento inicial, permitiendo determinar si la complejidad de los modelos no lineales es necesaria para mejorar la precisi√≥n. Es vital para la interpretabilidad de los coeficientes de las *features*, lo que facilita la comprensi√≥n de la influencia de cada variable en la predicci√≥n. Acad√©micamente, su inclusi√≥n es crucial para demostrar la mejora incremental que ofrecen modelos m√°s complejos y para validar la necesidad de t√©cnicas avanzadas cuando las relaciones lineales no son suficientes para capturar la complejidad de los datos [1].

#### 2. Random Forest Classifier (Bosque Aleatorio) üå≥

**Contribuci√≥n Metodol√≥gica:** El Random Forest representa una t√©cnica de ensamble por *bagging* (Bootstrap Aggregating). Se elige por su robustez, su baja sensibilidad al sobreajuste (*overfitting*) y su capacidad inherente para manejar la no linealidad en los datos. Su principal aporte es que permite la identificaci√≥n de la importancia de las *features*, mejorando la interpretabilidad del conjunto de predictores. La combinaci√≥n de m√∫ltiples √°rboles de decisi√≥n reduce la varianza y mejora la estabilidad del modelo, lo que lo convierte en una opci√≥n s√≥lida para problemas de clasificaci√≥n complejos y una excelente alternativa a los modelos lineales cuando la interpretabilidad a nivel de caracter√≠stica es importante [2].

#### 3. Gradient Boosting (Gradient Boosting Classifier) üöÄ

**Contribuci√≥n Metodol√≥gica:** El Gradient Boosting se incluye como una t√©cnica de ensamble secuencial (*boosting*). Su prop√≥sito es alcanzar el m√°ximo rendimiento predictivo corrigiendo iterativamente los errores de los √°rboles anteriores. Es un modelo clave en la fase de optimizaci√≥n (Hito 3) para determinar el l√≠mite superior de precisi√≥n alcanzable en el problema. Su capacidad para aprender de los errores residuales de modelos previos lo hace excepcionalmente potente para problemas donde se requiere una alta precisi√≥n, a menudo superando a otros algoritmos en competiciones de Machine Learning. Su inclusi√≥n permite explorar el potencial de mejora continua a trav√©s de la combinaci√≥n de modelos d√©biles [3].

#### 4. Perceptr√≥n Multicapa (MLP) / Red Neuronal üß†

**Contribuci√≥n Metodol√≥gica:** Su inclusi√≥n permite evaluar el rendimiento de un modelo de Deep Learning de entrada. El MLP es esencial para contrastar la efectividad del aprendizaje jer√°rquico de *features* (propio de las redes neuronales) frente a los m√©todos de ensamble cl√°sicos, lo cual es fundamental en la fase de Comparaci√≥n (Hito 4). A diferencia de la Regresi√≥n Log√≠stica (RL), las Redes Neuronales (RN) son capaces de aprender patrones y relaciones muy complejas y no lineales en los datos, lo que se revela gran importancia para la predicci√≥n de la variable objetivo, ya que las relaciones entre las variables X e Y podr√≠an ser muy sutiles. El aprendizaje de representaciones en las capas ocultas de una RN es capaz de aprender representaciones de caracter√≠sticas m√°s abstractas y √∫tiles a partir de los datos de entrada, lo que a veces elimina la necesidad de una ingenier√≠a exhaustiva de *features*. Adem√°s, en muchos problemas de clasificaci√≥n, las RN, especialmente cuando se ajustan correctamente y se dispone de suficientes datos, alcanzan un rendimiento superior al de los modelos tradicionales de Machine Learning, justificado as√≠ su exploraci√≥n como un candidato de alto rendimiento [4].

###  3. Preprocesamiento de Datos (Data Preprocessing).

#### 3.1 Explicaci√≥n Metodol√≥gica del C√≥digo de Preparaci√≥n

El c√≥digo escrito realiza los pasos fundamentales de Separaci√≥n de Datos y Estandarizaci√≥n de Features,  
procesos necesarios que deben ejecutarse rigurosamente antes de iniciar el entrenamiento de modelos de Machine Learning (ML).

1. Separaci√≥n de Variables (X e Y) y Divisi√≥n de Datos

Esta etapa define el problema de clasificaci√≥n y asegura la reproducibilidad del estudio:

Definici√≥n de Variables: Se a√≠sla la Variable Objetivo (Y = data['bloque_ideologico_encoded']), que el modelo intentar√° predecir,  
del conjunto de Variables Predictoras o Features (X), que son los insumos del modelo.

Divisi√≥n (Train-Test Split): Se emplea train_test_split para particionar el dataset en conjuntos de Entrenamiento (80%) y Prueba (20%).

Finalidad Acad√©mica: Reservar un conjunto de prueba (X_test, Y_test) permite una evaluaci√≥n final imparcial del rendimiento del modelo,  
garantizando que este sea validado sobre datos que nunca ha visto, previniendo as√≠ el sobreajuste (overfitting).

El uso de random_state=42 garantiza que esta partici√≥n sea reproducible en futuras ejecuciones.

2. Escalamiento de Features (StandardScaler)

El escalado es obligatorio para los modelos basados en la distancia o el gradiente.

Identificaci√≥n de Variables a Escalar: Las variables edad, ingreso_hogar, probabilidad_voto_generales, Renta_Per_Capita_2023_miles_euros  
y Renta_Per_Capita_2023_miles_euros_escalada son identificadas como variables num√©ricas con magnitudes o rangos dispares.

Aplicaci√≥n del StandardScaler:

Normalizaci√≥n: Este m√©todo transforma los datos de modo que la distribuci√≥n de cada feature tenga una media (Œº)  
de cero y una desviaci√≥n est√°ndar (œÉ) de uno.

Necesidad Algor√≠tmica: Modelos como la Regresi√≥n Log√≠stica, SVM y MLP son sensibles a la escala. Sin la estandarizaci√≥n,  
las features con rangos amplios (ej., la renta) dominar√≠an el c√°lculo de la funci√≥n de p√©rdida y la optimizaci√≥n del gradiente,  
sesgando el aprendizaje.

Prevenci√≥n de Data Leakage: Se utiliza fit_transform solo en el conjunto de entrenamiento (X_train) para que el escalador aprenda  
los par√°metros (Œº y œÉ) exclusivamente de esos datos. Luego, se utiliza solo transform en el conjunto de prueba (X_test) para aplicar  
la misma transformaci√≥n, evitando as√≠ la fuga de informaci√≥n (informaci√≥n del test set al train set).

```python
# 1. Definir la Variable Objetivo (Y) y las Features (X)
Y = data['bloque_ideologico_encoded']
X = data.drop(columns=['bloque_ideologico_encoded']) 

# 2. Separar datos de entrenamiento y prueba
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# 3. Identificar columnas num√©ricas para escalar
# Del CSV, estas son las que se tienen que escalar:
columnas_a_escalar = [
    'edad', 
    'ingreso_hogar',
    'probabilidad_voto_generales',
    'Renta_Per_Capita_2023_miles_euros', 
    'Renta_Per_Capita_2023_miles_euros_escalada' # Aunque ya est√° escalada, el StandardScaler es buena pr√°ctica.
]

# 4. Aplicar el StandardScaler SOLO a las columnas de entrenamiento
scaler = StandardScaler()
X_train[columnas_a_escalar] = scaler.fit_transform(X_train[columnas_a_escalar])

# 5. Aplicar la misma transformaci√≥n a los datos de prueba
X_test[columnas_a_escalar] = scaler.transform(X_test[columnas_a_escalar])
```

### 3.1.1 Se guardan los valores del Train y el Test para ser reutilizados

y no tener que volver a repetir la l√≠nea de c√≥digo anterior obteniendo as√≠ mas efciciencia.



```python
import joblib

# Los DataFrames X_train, X_test, Y_train y Y_test ya est√°n en la memoria,
# listos y escalados. Ahora los guardamos para el Notebook 3 (Hito 2).

joblib.dump(X_train, 'X_train.pkl')
joblib.dump(X_test, 'X_test.pkl')
joblib.dump(Y_train, 'Y_train.pkl')
joblib.dump(Y_test, 'Y_test.pkl')

print("Datos de entrenamiento y prueba guardados con √©xito en formato .pkl!")
```
