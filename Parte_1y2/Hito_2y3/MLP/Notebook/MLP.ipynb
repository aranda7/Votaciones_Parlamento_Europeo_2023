# Arquitectura de la MLP

````python
# Definimos la red neuronal
class MLP(nn.Module):
    """
    Perceptr√≥n Multicapa (MLP) configurable para clasificaci√≥n multiclase.

    Par√°metros:
    - input_size: n√∫mero de caracter√≠sticas de entrada
    - hidden_sizes: lista con tama√±o de cada capa oculta
    - output_size: n√∫mero de clases de salida
    - activation: funci√≥n de activaci√≥n ("relu", "tanh", "sigmoid", "leakyrelu")
    - dropout: porcentaje de dropout para regularizaci√≥n (0.0 significa sin dropout)
    """

    def __init__(self, input_size, hidden_sizes, output_size, activation="relu", dropout=0.0):
        super(MLP, self).__init__()

        # Diccionario de activaciones disponibles
        activations = {
            "relu": nn.ReLU(),
            "tanh": nn.Tanh(),
            "sigmoid": nn.Sigmoid(),
            "leakyrelu": nn.LeakyReLU(),
        }

        if activation.lower() not in activations:
            raise ValueError(f"Funci√≥n de activaci√≥n '{activation}' no soportada. "
                             f"Elige entre: {list(activations.keys())}")

        act_layer = activations[activation.lower()]

        layers = []
        in_dim = input_size

        # Capas ocultas con activaci√≥n seleccionada
        for h_dim in hidden_sizes:
            layers.append(nn.Linear(in_dim, h_dim))
            layers.append(act_layer)
            if dropout > 0:
                layers.append(nn.Dropout(dropout))
            in_dim = h_dim

        # Capa de salida (logits)
        layers.append(nn.Linear(in_dim, output_size))

        # Apilamos todas las capas
        self.network = nn.Sequential(*layers)

    # Forward pass
    def forward(self, x):
        return self.network(x)
````

### Arquitectura del Perceptr√≥n Multicapa (MLP)

La clase `MLP` implementa una red neuronal feedforward configurable para tareas de **clasificaci√≥n multiclase**.

#### Par√°metros principales
- **`input_size`**: n√∫mero de caracter√≠sticas de entrada del dataset.  
- **`hidden_sizes`**: lista con el n√∫mero de neuronas en cada capa oculta (ej: `[128, 64]`).  
- **`output_size`**: n√∫mero de clases de salida (logits).  
- **`activation`**: funci√≥n de activaci√≥n para las capas ocultas. Opciones:  
  - `"relu"` (por defecto)  
  - `"tanh"`  
  - `"sigmoid"`  
  - `"leakyrelu"`  
- **`dropout`**: porcentaje de desactivaci√≥n aleatoria de neuronas durante el entrenamiento (`0.0` significa sin dropout).

#### Flujo de la red
1. Se construyen din√°micamente las **capas ocultas** seg√∫n `hidden_sizes`.  
2. Despu√©s de cada capa lineal (`Linear`), se aplica la **funci√≥n de activaci√≥n** seleccionada.  
3. Opcionalmente, se a√±ade una **capa Dropout** para regularizaci√≥n.  
4. La **capa final** es un `Linear` que produce logits (sin activaci√≥n softmax, ya que `CrossEntropyLoss` la aplica internamente).  

#### Ventajas
- Arquitectura flexible y f√°cilmente ajustable para b√∫squeda de hiperpar√°metros.  
- Permite experimentar con diferentes funciones de activaci√≥n y profundidades.  
- Separa claramente la definici√≥n de la arquitectura del proceso de entrenamiento.

````python
# Ruta base
path = "/Users/daniel/Desktop/Master/TFM/Parte 2 TFM/MLP/"

# Cargar conjuntos balanceados de entrenamiento
X_train_oversampled = joblib.load(path + "X_train_oversampled.pkl")
Y_train_oversampled = joblib.load(path + "Y_train_oversampled.pkl")

# Cargar conjuntos originales de prueba
X_test = joblib.load(path + "X_test.pkl")
Y_test = joblib.load(path + "Y_test.pkl")

# Asegurar tipos correctos
X_train_oversampled = pd.DataFrame(X_train_oversampled)
X_test = pd.DataFrame(X_test)
Y_train_oversampled = pd.Series(Y_train_oversampled)
Y_test = pd.Series(Y_test)

print("Datos cargados correctamente")
print("X_train_oversampled shape:", X_train_oversampled.shape)
print("Y_train_oversampled shape:", Y_train_oversampled.shape)
print("X_test shape:", X_test.shape)
print("Y_test shape:", Y_test.shape)
````

# Creaci√≥n de los DataLoaders para entrenamiento y prueba

````python
#  Creaci√≥n de DataLoaders para entrenamiento y prueba
def get_loader(batch_size, X, y=None, shuffle=True, return_tensors=False):
    # Convertimos los datos a tensores de PyTorch
    X_tensor = torch.tensor(X.values, dtype=torch.float32).to(device)
    y_tensor = torch.tensor(y.values, dtype=torch.long).to(device)  # long para CrossEntropyLoss

    # Creamos TensorDataset y DataLoader
    dataset = TensorDataset(X_tensor, y_tensor)
    loader = DataLoader(dataset, batch_size=batch_size, shuffle=shuffle)

    if return_tensors:
        return loader, X_tensor, y_tensor
    else:
        return loader

#Crear DataLoaders para entrenamiento y prueba
batch_size = 64  # Preparado para ajustar
train_loader = get_loader(batch_size, X_train_oversampled, Y_train_oversampled)
test_loader = get_loader(batch_size, X_test, Y_test, shuffle=False)

print("DataLoaders preparados correctamente")
````

## **Creaci√≥n de DataLoaders para entrenamiento y prueba**
> En este bloque se preparan los conjuntos de datos para ser utilizados en el entrenamiento y la evaluaci√≥n de la MLP.  
> La preparaci√≥n incluye los siguientes pasos:

1. **Conversi√≥n a tensores de PyTorch:**  
   - Los datos de entrada `X` se convierten a tensores de tipo `float32`.  
   - Las etiquetas `y` se convierten a tensores de tipo `long` para ser compatibles con `CrossEntropyLoss`, que requiere enteros para clasificaci√≥n multiclase.  
   - Los tensores se mueven al dispositivo de c√≥mputo seleccionado (`CPU` o `GPU`).

2. **Creaci√≥n de TensorDataset y DataLoader:**  
   - `TensorDataset` combina los tensores de caracter√≠sticas y etiquetas en un √∫nico objeto iterable.  
   - `DataLoader` permite iterar sobre los datos en **batches**, facilitando:
     - Entrenamiento por mini-batches.
     - Shuffle de los datos durante entrenamiento para mejorar la generalizaci√≥n.

3. **Configuraci√≥n de batches:**  
   - Se define un tama√±o de batch de `64`, que puede ajustarse seg√∫n memoria y eficiencia.  
   - El DataLoader de test (`test_loader`) no aplica shuffle para mantener el orden de evaluaci√≥n.

4. **Objetivo:**  
   - Separar la preparaci√≥n de los datos del entrenamiento del modelo.  
   - Garantizar un flujo modular y reproducible, permitiendo reutilizar los DataLoaders con diferentes modelos sin modificar la preparaci√≥n de datos.

**Conclusi√≥n:**  
Este bloque asegura que los datos est√©n correctamente formateados y organizados para el entrenamiento de la MLP, promoviendo eficiencia y claridad en el flujo de trabajo.


# Instanciar la MLP

````python
input_size = X_train_oversampled.shape[1]  # N√∫mero de features de entrada (22 columnas)

mlp_model = MLP(
    input_size=input_size,
    hidden_sizes=[32, 16],           # ejemplo de capas ocultas
    output_size=Y_train_oversampled.nunique(),  # n√∫mero de clases
    dropout=0.2
)

mlp_model = mlp_model.to(device)
print(mlp_model)
````

  - `output_size = Y_train_oversampled.nunique()` ‚Üí n√∫mero de clases del problema (8 clases).  
  - `dropout = 0.2` ‚Üí tasa de dropout para regularizaci√≥n.  

- La funci√≥n de activaci√≥n de las capas ocultas es **ReLU**, consistente con la consigna del Hito 3.

```python
mlp_model = mlp_model.to(device)
```

# Entrenamiento

````python
# Definir la funci√≥n de p√©rdida para multiclase
criterion = nn.CrossEntropyLoss()

# Definir el optimizador
optimizer = optim.Adam(mlp_model.parameters(), lr=0.001)

# N√∫mero de epochs
epochs = 30

# Historial de p√©rdidas
loss_history = {
    "train_loss": [],
    "val_loss": []
}

# Entrenamiento
for epoch in tqdm(range(1, epochs + 1), desc="Entrenando..."):
    mlp_model.train()
    total_train_loss = 0

    # Bucle por batches
    for x_batch, y_batch in train_loader:
        x_batch, y_batch = x_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        logits = mlp_model(x_batch)
        loss = criterion(logits, y_batch)
        loss.backward()
        optimizer.step()
        total_train_loss += loss.item() * x_batch.size(0)

    avg_train_loss = total_train_loss / len(train_loader.dataset)
    loss_history["train_loss"].append(avg_train_loss)

    # Validaci√≥n
    mlp_model.eval()
    total_val_loss = 0
    with torch.no_grad():
        for x_val, y_val in test_loader:
            x_val, y_val = x_val.to(device), y_val.to(device)
            val_logits = mlp_model(x_val)
            val_loss = criterion(val_logits, y_val)
            total_val_loss += val_loss.item() * x_val.size(0)
    avg_val_loss = total_val_loss / len(test_loader.dataset)
    loss_history["val_loss"].append(avg_val_loss)

    # Mostrar progreso
    tqdm.write(f"Epoch {epoch}/{epochs} - Train Loss: {avg_train_loss:.4f} - Val Loss: {avg_val_loss:.4f}")

print("Entrenamiento completado")
````

- **Test original**, reflejando la distribuci√≥n real de clases (Clase 6 ausente en test).

### 1. Wrapper para Modelo Default



````python
class MLPWrapperDefault:
    """Wrapper para hacer el modelo PyTorch compatible con permutation_importance"""
    
    def __init__(self, model, device):
        self.model = model
        self.device = device
        self.model.eval()
    
    def fit(self, X, y):
        """M√©todo fit requerido por scikit-learn"""
        return self
    
    def predict(self, X):
        """M√©todo predict compatible con scikit-learn"""
        self.model.eval()
        with torch.no_grad():
            if not isinstance(X, torch.Tensor):
                X_tensor = torch.tensor(X, dtype=torch.float32).to(self.device)
            else:
                X_tensor = X.to(self.device)
            
            outputs = self.model(X_tensor)
            _, predicted = torch.max(outputs, 1)
            
            # Usar tolist() como has estado haciendo
            predictions_list = predicted.cpu().tolist()
            
            # Convertir a numpy array para compatibilidad con scikit-learn
            import numpy as np
            return np.array(predictions_list)

print("‚úì Wrapper para modelo default creado")
````

### 2. C√°lculo de Feature Importance Default

````python
print("C√°lculo Feature Importance para Modelo Default...")

# Crear wrapper del modelo default (usando 'mlp_model')
wrapper_default = MLPWrapperDefault(mlp_model, device)

# Calcular accuracy baseline
y_pred_default = wrapper_default.predict(X_test.values)
baseline_accuracy_default = accuracy_score(Y_test, y_pred_default)
print(f"Accuracy (Default): {baseline_accuracy_default:.4f}")

# Calcular permutation importance
perm_importance_default = permutation_importance(
    wrapper_default, 
    X_test.values, 
    Y_test, 
    n_repeats=10,
    random_state=42,
    scoring='accuracy'
)

print("‚úì Permutation importance calculado para modelo default")
````

#### Justificaci√≥n Metodol√≥gica: *MLPWrapper*

La implementaci√≥n del **MLPWrapper** constituye una soluci√≥n de ingenier√≠a necesaria para la integraci√≥n entre frameworks de *deep learning* (**PyTorch**) y herramientas de interpretabilidad de *machine learning* tradicional (**scikit-learn**).  

Esta adaptaci√≥n es metodol√≥gicamente esencial por las siguientes razones:

1. **Compatibilidad de Interfaces**  
   Los modelos PyTorch no implementan nativamente la interfaz est√°ndar de scikit-learn (`fit`, `predict`), requerida por `permutation_importance`.  
   El wrapper act√∫a como un *adapter pattern* que traduce las llamadas de scikit-learn a operaciones PyTorch.

2. **Gesti√≥n de Estados del Modelo**  
   El wrapper garantiza que el modelo permanezca en modo evaluaci√≥n (`model.eval()`) durante las predicciones, desactivando componentes como *dropout* y *batch normalization* que alterar√≠an los resultados de importancia.

3. **Manejo de Tensores**  
   Facilita la conversi√≥n autom√°tica entre arrays NumPy (utilizados por scikit-learn) y tensores PyTorch, incluyendo la gesti√≥n apropiada de dispositivos (CPU/GPU) y tipos de datos.



#### Conclusi√≥n

Esta aproximaci√≥n mantiene la **integridad metodol√≥gica** del an√°lisis de interpretabilidad mientras preserva la arquitectura original del modelo neuronal, asegurando que los resultados de *Feature Importance* reflejen fielmente el comportamiento del modelo entrenado.

# Curvas ROC Multiclase

````python
# N√∫mero de clases
n_classes = Y_train_oversampled.nunique()

# Binarizar etiquetas del test (one-vs-rest)
y_test_bin = label_binarize(Y_test, classes=np.arange(n_classes))

# Predicciones probabil√≠sticas con softmax (sin .numpy())
with torch.no_grad():
    logits = mlp_model(torch.tensor(X_test.values, dtype=torch.float32).to(device))
    y_pred_proba = torch.softmax(logits, dim=1).detach().cpu()

# Inicializar diccionarios para ROC y AUC
fpr = {}
tpr = {}
roc_auc = {}

# ROC y AUC por clase
for i in range(n_classes):
    y_true_i = y_test_bin[:, i].tolist()
    y_score_i = y_pred_proba[:, i].tolist()
    fpr[i], tpr[i], _ = roc_curve(y_true_i, y_score_i)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro-average
fpr["micro"], tpr["micro"], _ = roc_curve(y_test_bin.ravel().tolist(), y_pred_proba.ravel().tolist())
roc_auc["micro"] = auc(fpr["micro"], tpr["micro"])

# Macro-average
all_fpr = np.unique(np.concatenate([fpr[i] for i in range(n_classes)]))
mean_tpr = np.zeros_like(all_fpr)
for i in range(n_classes):
    mean_tpr += np.interp(all_fpr, fpr[i], tpr[i])
mean_tpr /= n_classes
fpr["macro"], tpr["macro"] = all_fpr, mean_tpr
roc_auc["macro"] = auc(fpr["macro"], tpr["macro"])

# Crear DataFrame con FPR y TPR de cada clase, micro y macro
roc_data = {}
for i in range(n_classes):
    roc_data[f'Clase {i} FPR'] = fpr[i]
    roc_data[f'Clase {i} TPR'] = tpr[i]
roc_data['Micro FPR'] = fpr['micro']
roc_data['Micro TPR'] = tpr['micro']
roc_data['Macro FPR'] = fpr['macro']
roc_data['Macro TPR'] = tpr['macro']

roc_df = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in roc_data.items()]))

# Mostrar DataFrame
print(roc_df)

# Graficar ROC
plt.figure(figsize=(10,7))
plt.plot(fpr["micro"], tpr["micro"], label=f'Micro-average ROC (AUC = {roc_auc["micro"]:.2f})', color='deeppink', linestyle=':', linewidth=4)
plt.plot(fpr["macro"], tpr["macro"], label=f'Macro-average ROC (AUC = {roc_auc["macro"]:.2f})', color='navy', linestyle=':', linewidth=4)

colors = plt.cm.tab10(np.linspace(0, 1, n_classes))
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Curvas ROC Multiclase - MLP (One-vs-Rest)')
plt.legend(loc="lower right")
plt.show()
````

# Funci√≥n de entrenamiento y evaluaci√≥n 

````python
def train_and_evaluate(params, X_train_oversampled, Y_train_oversampled, X_test, Y_test,
                       device=device, patience=5):
    """
    Entrena y eval√∫a una MLP en funci√≥n de los hiperpar√°metros recibidos.
    Incluye early stopping basado en la p√©rdida de validaci√≥n.

    Par√°metros:
    - params: diccionario con hiperpar√°metros (hidden_sizes, activation, dropout,
              batch_size, lr, optimizer, epochs)
    - X_train_oversampled, Y_train_oversampled: datos de entrenamiento balanceados
    - X_test, Y_test: datos de prueba originales
    - device: 'cuda' o 'cpu'
    - patience: n√∫mero de √©pocas sin mejora antes de detener el entrenamiento

    Retorna:
    - m√©tricas (dict)
    - historial de p√©rdidas (dict)
    - modelo entrenado (nn.Module)
    """

    # Crear DataLoaders
    train_loader = get_loader(params["batch_size"], X_train_oversampled, Y_train_oversampled)
    test_loader = get_loader(params["batch_size"], X_test, Y_test, shuffle=False)

    # Definir modelo
    model = MLP(
        input_size=X_train_oversampled.shape[1],
        hidden_sizes=params["hidden_sizes"],
        output_size=Y_train_oversampled.nunique(),
        activation=params["activation"],
        dropout=params["dropout"]
    ).to(device)

    # Criterio
    criterion = nn.CrossEntropyLoss()

    # Optimizer
    if params["optimizer"].lower() == "adam":
        optimizer = optim.Adam(model.parameters(), lr=params["lr"])
    elif params["optimizer"].lower() == "sgd":
        optimizer = optim.SGD(model.parameters(), lr=params["lr"], momentum=0.9)
    elif params["optimizer"].lower() == "rmsprop":
        optimizer = optim.RMSprop(model.parameters(), lr=params["lr"])
    else:
        raise ValueError("Optimizador no soportado")

    # Historial de p√©rdidas
    loss_history = {"train_loss": [], "val_loss": []}

    # Early stopping
    best_val_loss = float('inf')
    epochs_no_improve = 0

    # Bucle de entrenamiento
    for epoch in range(1, params["epochs"] + 1):
        model.train()
        total_train_loss = 0
        for x_batch, y_batch in train_loader:
            x_batch, y_batch = x_batch.to(device), y_batch.to(device)
            optimizer.zero_grad()
            logits = model(x_batch)
            loss = criterion(logits, y_batch)
            loss.backward()
            optimizer.step()
            total_train_loss += loss.item() * x_batch.size(0)

        avg_train_loss = total_train_loss / len(train_loader.dataset)
        loss_history["train_loss"].append(avg_train_loss)

        # Validaci√≥n
        model.eval()
        total_val_loss = 0
        y_true, y_pred = [], []
        with torch.no_grad():
            for x_val, y_val in test_loader:
                x_val, y_val = x_val.to(device), y_val.to(device)
                val_logits = model(x_val)
                val_loss = criterion(val_logits, y_val)
                total_val_loss += val_loss.item() * x_val.size(0)
                preds = torch.argmax(val_logits, dim=1)
                y_true.extend(y_val.cpu().tolist())
                y_pred.extend(preds.cpu().tolist())

        avg_val_loss = total_val_loss / len(test_loader.dataset)
        loss_history["val_loss"].append(avg_val_loss)

        # ‚úÖ Print de progreso por √©poca
        print(f"üìò Epoch {epoch}/{params['epochs']} - "
              f"Train Loss: {avg_train_loss:.4f} | Val Loss: {avg_val_loss:.4f}")

       # --- Early stopping ---
        if avg_val_loss < best_val_loss:
            print(f"‚úÖ ¬°MEJORA! Val Loss: {best_val_loss:.4f} ‚Üí {avg_val_loss:.4f}")
            best_val_loss = avg_val_loss
            epochs_no_improve = 0
            best_model_state = model.state_dict()
        else:
            epochs_no_improve += 1
            print(f"‚ùå Sin mejora. Actual: {avg_val_loss:.4f}, Mejor: {best_val_loss:.4f} ({epochs_no_improve}/{patience})")

            if epochs_no_improve >= patience:
                print(f"‚èπÔ∏è Early stopping en epoch {epoch} (no mejora en {patience} √©pocas)")
                model.load_state_dict(best_model_state)
                break

    # M√©tricas finales
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average="weighted", zero_division=0)
    recall = recall_score(y_true, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)

    metricas = {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "classification_report": classification_report(y_true, y_pred, digits=4, zero_division=0)
    }

    print("\nüéØ M√©tricas finales:")
    print(f"Accuracy: {accuracy:.4f} | Precision: {precision:.4f} | "
          f"Recall: {recall:.4f} | F1-score: {f1:.4f}")

    return metricas, loss_history, model






# Imprimir informaci√≥n pertinente


print(f"\nüìà M√âTRICAS DE EVALUACI√ìN:")
print(f"  ‚Ä¢ Accuracy: {metricas['accuracy']:.4f}")
print(f"  ‚Ä¢ Precision: {metricas['precision']:.4f}")
print(f"  ‚Ä¢ Recall: {metricas['recall']:.4f}")
print(f"  ‚Ä¢ F1-Score: {metricas['f1']:.4f}")

print(f"\nüìâ HISTORIAL DE P√âRDIDAS:")
print(f"  ‚Ä¢ P√©rdida final de entrenamiento: {loss_history['train_loss'][-1]:.4f}")
print(f"  ‚Ä¢ P√©rdida final de validaci√≥n: {loss_history['val_loss'][-1]:.4f}")
print(f"  ‚Ä¢ Total de √©pocas ejecutadas: {len(loss_history['train_loss'])}")

print(f"\nüìã CLASSIFICATION REPORT:")
print(metricas['classification_report'])


````

# Random Search de hiperpar√°metros

````python
def random_search(param_space, n_iter=10, X_train=X_train_oversampled, Y_train=Y_train_oversampled,
                  X_test=X_test, Y_test=Y_test, device=device, random_seed=42):

    random.seed(random_seed)
    best_score = 0
    best_params = None
    best_model = None
    results = []

    # Obtener todas las combinaciones posibles (sin repetir)
    for i in range(n_iter):
        # Elegir aleatoriamente un valor de cada hiperpar√°metro
        params = {k: random.choice(v) for k, v in param_space.items()}
        print(f"\n Iteraci√≥n {i+1}/{n_iter} - Probando par√°metros: {params}")

        # Entrenar y evaluar
        metricas, loss_history, model = train_and_evaluate(
            params=params,
            X_train_oversampled=X_train,
            Y_train_oversampled=Y_train,
            X_test=X_test,
            Y_test=Y_test,
            device=device
        )

        # Guardar resultados
        results.append({
            "params": params,
            "metricas": metricas,
            "loss_history": loss_history
        })

        # Actualizar mejor modelo seg√∫n F1 ponderado
        if metricas["f1"] > best_score:
            best_score = metricas["f1"]
            best_params = params
            best_model = model

        print(f" F1-score ponderado actual: {metricas['f1']:.4f}")

    print("\nüéØ Mejor combinaci√≥n encontrada:")
    print(best_params)
    print(f"F1-score ponderado: {best_score:.4f}")

    return best_model, best_params, results

    # Ejecutar Random Search con, por ejemplo, 5 combinaciones
best_model, best_params, results = random_search(param_space, n_iter=5)

# inspeccionar resultados
print("\nMejor configuraci√≥n de hiperpar√°metros:")
print(best_params)

print("\nF1-score ponderado del mejor modelo:")
print(max(r["metricas"]["f1"] for r in results))

````

# Feature Importance para Modelo Optimizado (Hyperparams)

### Bloque B1: Wrapper para Modelo Optimizado

````python
# WRAPPER PARA MODELO OPTIMIZADO - FEATURE IMPORTANCE
class MLPWrapperHyperparams:
    """Wrapper para hacer el modelo PyTorch optimizado compatible con permutation_importance"""
    
    def __init__(self, model, device):
        self.model = model
        self.device = device
        self.model.eval()
    
    def predict(self, X):
        """M√©todo predict compatible con scikit-learn"""
        self.model.eval()
        with torch.no_grad():
            if isinstance(X, torch.Tensor):
                # Si ya es tensor -> enviamos al device y casteamos
                X_tensor = X.to(self.device, dtype=torch.float32)
            else:
                # Si es NumPy array o lista -> convertimos con .tolist()
                X_tensor = torch.tensor(
                    X.tolist() if hasattr(X, "tolist") else X,
                    dtype=torch.float32
                ).to(self.device)
            
            outputs = self.model(X_tensor)
            _, predicted = torch.max(outputs, 1)
            return predicted.cpu().numpy()

print("‚úì Wrapper para modelo optimizado creado con .tolist()")
````

````python
# N√∫mero de clases
n_classes = Y_train_oversampled.nunique()

# Binarizar etiquetas de test
y_test_bin = label_binarize(Y_test, classes=list(range(n_classes)))

# Probabilidades predichas por el modelo (tensor -> lista)
with torch.no_grad():
    y_pred_proba_tensor = best_model(torch.tensor(X_test.values, dtype=torch.float32).to(device))
y_pred_proba = y_pred_proba_tensor.detach().cpu().tolist()

# Inicializar diccionarios
fpr = {}
tpr = {}
roc_auc = {}

# Curvas por clase
for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve([row[i] for row in y_test_bin], [row[i] for row in y_pred_proba])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Micro y Macro AUC
roc_auc["micro"] = roc_auc_score(y_test_bin, y_pred_proba, average="micro")
roc_auc["macro"] = roc_auc_score(y_test_bin, y_pred_proba, average="macro")

# FPR y TPR micro
fpr["micro"], tpr["micro"], _ = roc_curve(
    [v for row in y_test_bin for v in row], 
    [v for row in y_pred_proba for v in row]
)

# FPR y TPR macro: promedio de todas las clases
all_fpr = sorted(set(sum([list(fpr[i]) for i in range(n_classes)], [])))
mean_tpr = [0.0]*len(all_fpr)
for i in range(n_classes):
    interp_tpr = np.interp(all_fpr, list(fpr[i]), list(tpr[i]))
    mean_tpr = [sum(x) for x in zip(mean_tpr, interp_tpr)]
mean_tpr = [x / n_classes for x in mean_tpr]
fpr["macro"] = all_fpr
tpr["macro"] = mean_tpr

# Crear DataFrame usando listas puras
roc_dict = {}
for i in range(n_classes):
    roc_dict[f'FPR_Clase_{i}'] = list(fpr[i])
    roc_dict[f'TPR_Clase_{i}'] = list(tpr[i])
roc_dict['FPR_micro'] = list(fpr['micro'])
roc_dict['TPR_micro'] = list(tpr['micro'])
roc_dict['FPR_macro'] = list(fpr['macro'])
roc_dict['TPR_macro'] = list(tpr['macro'])

# Normalizar longitudes para DataFrame
max_len = max(len(v) for v in roc_dict.values())
for k in roc_dict:
    roc_dict[k] = roc_dict[k] + [None]*(max_len - len(roc_dict[k]))

roc_df = pd.DataFrame(roc_dict)
print("=== DataFrame de Curvas ROC ===")
print(roc_df)

# Plot
plt.figure(figsize=(10,7))
plt.plot(fpr["micro"], tpr["micro"], label=f'Micro-average ROC (AUC={roc_auc["micro"]:.2f})', color='deeppink', linestyle=':', linewidth=4)
plt.plot(fpr["macro"], tpr["macro"], label=f'Macro-average ROC (AUC={roc_auc["macro"]:.2f})', color='navy', linestyle=':', linewidth=4)

colors = plt.cm.tab10(range(n_classes))
for i, color in zip(range(n_classes), colors):
    plt.plot(fpr[i], tpr[i], color=color, lw=2, label=f'Clase {i} (AUC={roc_auc[i]:.2f})')

plt.plot([0,1],[0,1],'k--', lw=2)
plt.xlim([0,1])
plt.ylim([0,1.05])
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Curvas ROC Multiclase - MLP Optimizado')
plt.legend(loc='lower right')
plt.show()
````

#### Investigaci√≥n: Significado de micro y macro en Curvas ROC multiclase

En problemas de clasificaci√≥n multiclase, existen varias formas de resumir la capacidad de discriminaci√≥n del modelo usando las curvas ROC y el AUC:


‚àö 1. Micro-average ROC/AUC

- **Definici√≥n**: Trata todas las clases como si fueran una sola, **aplanando las predicciones** y las etiquetas verdaderas.  
- **C√≥mo se calcula**:  
  1. Se concatenan todas las instancias de todas las clases.  
  2. Se calcula la TPR y FPR global sobre este conjunto unificado.  
- **Interpretaci√≥n**:  
  - Refleja el **desempe√±o global del modelo** considerando cada ejemplo por igual, independientemente de su clase.  
  - Es sensible a clases dominantes, porque las clases con m√°s instancias contribuyen m√°s al c√°lculo.


‚àö 2. Macro-average ROC/AUC

- **Definici√≥n**: Calcula el AUC para cada clase por separado y luego hace un **promedio simple** de todas las clases.  
- **C√≥mo se calcula**:  
  1. Se obtiene la curva ROC y el AUC individual de cada clase (tratando esa clase como positiva y las dem√°s como negativas).  
  2. Se promedian todos los AUC para obtener el AUC macro.  
- **Interpretaci√≥n**:  
  - Eval√∫a el **rendimiento promedio entre clases**, dando la misma importancia a clases minoritarias y mayoritarias.  
  - Permite identificar si el modelo funciona bien de manera equilibrada, sin estar sesgado hacia clases con m√°s ejemplos.


‚àö 3. Resumen conceptual

| Tipo       | Considera | Sensible a | Uso principal |
|------------|-----------|------------|---------------|
| **Micro**  | Todas las instancias como un solo conjunto | Clases dominantes | Evaluaci√≥n global del modelo |
| **Macro**  | Cada clase por separado y luego promedio | Todas las clases por igual | Equilibrio y rendimiento por clase |

> En este contexto, **micro** nos dice cu√°n bien predice el modelo en general, mientras que **macro** nos dice si el modelo es **consistente entre todas las clases**, incluyendo las minoritarias.
